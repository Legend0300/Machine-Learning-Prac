```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("carprices.csv")
df.head()
```

```python
encoder = OneHotEncoder()
```

```python
encoded_df = encoder.fit_transform(df[['Car Model']])
```

```python
encoded_df = pd.DataFrame(encoded_df.toarray(), columns=encoder.get_feature_names_out(['Car Model']))
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df = df.iloc[1: , 1 : ]
df
```

```python

y = df["Sell Price($)"]

df.drop("Sell Price($)", axis=1, inplace=True)
df.drop("Car Model_Audi A5", axis=1, inplace=True)

# Now, you can directly use the modified DataFrame `df` without the "Sell Price($)" column
X = df
X
```

```python
model = linear_model.LinearRegression()
model.fit(X , y)
```

```python
pr1 = [45000 , 4  , 0 , 1]
pr2 = [86000 , 7  ,  1, 0]

print(model.predict([pr1 , pr2]))
model.score(X , y)
```
